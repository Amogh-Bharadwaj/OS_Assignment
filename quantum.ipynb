```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data/timequantum.csv")

df.rename(columns = {'c1waitq1': 'C1 WT at time quantum 1 sec', 'c2waitq1': 'C2 WT at time quantum 1 sec', 'c3waitq1': 'C3 WT at time quantum 1 sec'
                  ,'c1waitq2': 'C1 WT at time quantum 2 sec', 'c2waitq2': 'C2 WT at time quantum 2 sec', 'c3waitq2': 'C3 WT at time quantum 2 sec'   
                ,'c1waitq3': 'C1 WT at time quantum 3 sec', 'c2waitq3': 'C2 WT at time quantum 3 sec', 'c3waitq3': 'C3 WT at time quantum 3 sec'  
                , 'c1turnq1': 'C1 TAT at time quantum 1 sec', 'c2turnq1': 'C2 TAT at time quantum 1 sec', 'c3turnq1': 'C3 TAT at time quantum 1 sec'
          , 'c1turnq2': 'C1 TAT at time quantum 2 sec', 'c2turnq2': 'C2 TAT at time quantum 2 sec', 'c3turnq2': 'C3 TAT at time quantum 2 sec'
          , 'c1turnq3': 'C1 TAT at time quantum 3 sec', 'c2turnq3': 'C2 TAT at time quantum 3 sec', 'c3turnq3': 'C3 TAT at time quantum 3 sec'}
          , inplace = True)
df = df.set_index('n')
df.head()
```

```python
df1 = df[['C1 TAT at time quantum 1 sec','C1 TAT at time quantum 2 sec','C1 TAT at time quantum 3 sec']]
plt.figure()
df1.plot(title='C1 Turnaround Time at different time quantums')
plt.xlabel('Workload Size')
plt.ylabel('Time taken in s')
plt.show()
```
