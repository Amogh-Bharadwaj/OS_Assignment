```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data/rrdata1.csv")
df = df.set_index("n1=n2=n3")
df.rename(columns = {'c1wait': 'C1 Waiting Time', 'c2wait': 'C2 Waiting Time', 'c3wait': 'C3 Waiting Time'
                    , 'c1turn': 'C1 Turnaround Time', 'c2turn': 'C2 Turnaround Time', 'c3turn': 'C3 Turnaround Time'}, inplace = True)
df.head()
```

```python
df1 = df[['C1 Waiting Time','C1 Turnaround Time']]
plt.figure()
df1.plot(title='Waiting and Turnaround time for process C1 for time quantum = 1 sec')
plt.xlabel('Workload Size of C1')
plt.ylabel('Time taken in ms')
plt.show()
```

```python
df2 = df[['C2 Waiting Time','C2 Turnaround Time']]
plt.figure()
df2.plot(title='Waiting and Turnaround time for process C2 for time quantum = 1 sec')
plt.xlabel('Workload Size of C2')
plt.ylabel('Time taken in ms')
plt.show()
```

```python
df3 = df[['C3 Waiting Time','C3 Turnaround Time']]
plt.figure()
df3.plot(title='Waiting and Turnaround time for process C3 for time quantum = 1 sec')
plt.xlabel('Workload Size of C3')
plt.ylabel('Time taken in ms')
plt.show()
```
